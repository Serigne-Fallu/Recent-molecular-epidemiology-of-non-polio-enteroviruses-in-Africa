```python
import pandas as pd

# Chargement du fichier CSV
csv_path = "NPEV_ABCD_clean_dated.csv"
df = pd.read_csv(csv_path, sep=';')

# Fonction pour convertir les dates au format 'YYYY/MM/DD'
def convert_date_format(date):
    try:
        # Tentative de conversion des dates 'YYYY-MM-DD'
        date_converted = pd.to_datetime(date, format='%Y-%m')
    except ValueError:
        try:
            # Tentative de conversion des dates 'DD/MM/YYYY'
            date_converted = pd.to_datetime(date, format='%d/%m/%Y')
        except ValueError:
            return date  # Retourne la date telle quelle si les deux conversions échouent
    
    return date_converted.strftime('%Y/%m/%d')

# Application de la fonction de conversion à la colonne 'collection_date'
df['standardized_date'] = df['standardized_date'].apply(convert_date_format)

# Affichage du dataframe après transformation
print("Après transformation:")
print(df.head())

# Enregistrement du dataframe transformé dans un nouveau fichier CSV
output_path = "NPEV_ABCD_clean_datedYYYMMDD.csv"
df.to_csv(output_path, sep=';', index=False)
```
